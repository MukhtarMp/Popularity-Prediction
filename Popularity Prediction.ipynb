```python
# Calculate skewness and kurtosis for numerical columns
skewness = df[numerical_columns].skew()
kurtosis = df[numerical_columns].kurt()

print("Skewness:\n", skewness)
print("Kurtosis:\n", kurtosis)
```

<div style="background-color:#f0f8ff; padding:15px; border-left: 6px solid #008080; border-radius:5px;">
    <p style="color:#333333; font-size:18px; line-height:1.6;">
        The skewness and kurtosis values describe the distribution of the variables: <b>views</b>, <b>likes</b>, <b>dislikes</b>, and <b>comment_count</b>. Let’s break down these two measures:
    </p>
    <p style="color:#333333; font-size:18px; line-height:1.6;">
        <b>Skewness</b> measures the asymmetry of a distribution:
        <ul>
            <li><b>Positive skewness:</b> The right tail is longer, meaning most data is concentrated on the left with outliers on the right.</li>
            <li><b>Negative skewness:</b> The left tail is longer, meaning most data is on the right with outliers on the left.</li>
            <li>A skewness of <b>0</b> indicates a perfectly symmetrical distribution.</li>
        </ul>
    </p>
    <p style="color:#333333; font-size:18px; line-height:1.6;">
        <b>Interpretation of Skewness Values:</b>
        <ul>
            <li><b>Views (12.236):</b> Highly positive skewness, indicating most videos have low views, but a few have extremely high views (outliers).</li>
            <li><b>Likes (10.92):</b> Similar pattern—most videos have low likes, with a few receiving exceptionally high likes.</li>
            <li><b>Dislikes (40.17):</b> Very high skewness, where most videos have few dislikes, but some have an unusually large number of dislikes.</li>
            <li><b>Comment Count (19.74):</b> Highly skewed, with most videos having few comments, but some receiving disproportionately large numbers of comments.</li>
        </ul>
    </p>
    <p style="color:#333333; font-size:18px; line-height:1.6;">
        <b>Kurtosis</b> measures the "tailedness" of a distribution:
        <ul>
            <li><b>High kurtosis (&gt;3):</b> Indicates a distribution with heavy tails (more outliers).</li>
            <li><b>Low kurtosis (&lt;3):</b> Indicates a distribution with light tails (fewer outliers).</li>
            <li>A kurtosis of <b>3</b> corresponds to a normal distribution (mesokurtic).</li>
        </ul>
    </p>
    <p style="color:#333333; font-size:18px; line-height:1.6;">
        <b>Interpretation of Kurtosis Values:</b>
        <ul>
            <li><b>Views (232.20):</b> Extremely high kurtosis, suggesting a large number of outliers—few videos receive exceptionally high views.</li>
            <li><b>Likes (177.70):</b> High kurtosis indicates that most videos have relatively low likes, but some videos receive disproportionately high likes.</li>
            <li><b>Dislikes (1985.22):</b> Very high kurtosis shows many extreme outliers—some videos receive a huge number of dislikes.</li>
            <li><b>Comment Count (531.57):</b> Similar to likes and dislikes, there are many outliers with a few videos receiving an exceptionally large number of comments.</li>
        </ul>
    </p>
    <p style="color:#333333; font-size:18px; line-height:1.6;">
        <b>Overall Explanation:</b>
        <ul>
            <li>Both <b>skewness</b> and <b>kurtosis</b> values are very high, indicating significant positive skewness and heavy tails in the distribution.</li>
            <li>Most videos have low views, likes, dislikes, and comments, while a few outliers receive disproportionately high engagement.</li>
            <li>This extreme variability suggests the presence of outliers, which can be challenging for modeling without transformation or special handling.</li>
        </ul>

```python
def count_outliers_iqr(df_cleaned, exclude_columns=None):
    outlier_count = 0  # Initialize counter for total outliers

    if exclude_columns is None:
        exclude_columns = []  # If no columns are specified, initialize an empty list

    for col in df_cleaned.columns:
        if np.issubdtype(df_cleaned[col].dtype, np.number) and col not in exclude_columns:  # Check only numeric columns
            Q1 = np.percentile(df_cleaned[col], 25)  # 25th percentile (Q1)
            Q3 = np.percentile(df_cleaned[col], 75)  # 75th percentile (Q3)
            IQR = Q3 - Q1  # Interquartile range
            
            # Define outlier bounds
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            
            # Identify outliers
            outliers = df_cleaned[(df_cleaned[col] < lower_bound) | (df_cleaned[col] > upper_bound)]
            num_outliers = outliers.shape[0]
            outlier_count += num_outliers
            
            print(f"Feature: {col} | Outliers: {num_outliers}")
    
    return outlier_count

# Exclude 'views'and boolean columns while checking for outliers
total_outliers = count_outliers_iqr(df_cleaned, exclude_columns=['comments_disabled', 'ratings_disabled', 'video_error_or_removed','category_id'])

print(f"\nTotal number of outliers in the dataset: {total_outliers}")
```

```python
def cap_outliers(df_cleaned, columns):
    for col in columns:
        if np.issubdtype(df_cleaned[col].dtype, np.number):  # Check if the column is numeric
            Q1 = np.percentile(df_cleaned[col], 25)  # 25th percentile (Q1)
            Q3 = np.percentile(df_cleaned[col], 75)  # 75th percentile (Q3)
            IQR = Q3 - Q1  # Interquartile range
            
            # Define outlier bounds
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            
            # Cap the outliers
            df_cleaned[col] = np.where(df_cleaned[col] < lower_bound, lower_bound, df_cleaned[col])
            df_cleaned[col] = np.where(df_cleaned[col] > upper_bound, upper_bound, df_cleaned[col])
            
            print(f"Capped outliers for feature: {col}")

# Specify the columns to cap
columns_to_cap = ['views','likes', 'dislikes', 'comment_count']

# Cap the outliers in the specified columns
cap_outliers(df_cleaned, columns_to_cap)
```

```python
def count_outliers_iqr(unseen_df, exclude_columns=None):
    outlier_count = 0  # Initialize counter for total outliers

    if exclude_columns is None:
        exclude_columns = []  # If no columns are specified, initialize an empty list

    for col in unseen_df.columns:
        if np.issubdtype(unseen_df[col].dtype, np.number) and col not in exclude_columns:  # Check only numeric columns
            Q1 = np.percentile(unseen_df[col], 25)  # 25th percentile (Q1)
            Q3 = np.percentile(unseen_df[col], 75)  # 75th percentile (Q3)
            IQR = Q3 - Q1  # Interquartile range
            
            # Define outlier bounds
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            
            # Identify outliers
            outliers = unseen_df[(unseen_df[col] < lower_bound) | (unseen_df[col] > upper_bound)]
            num_outliers = outliers.shape[0]
            outlier_count += num_outliers
            
            print(f"Feature: {col} | Outliers: {num_outliers}")
    
    return outlier_count

# Exclude 'views'and boolean columns while checking for outliers
total_outliers = count_outliers_iqr(unseen_df, exclude_columns=['comments_disabled', 'ratings_disabled', 'video_error_or_removed','category_id'])

print(f"\nTotal number of outliers in the dataset: {total_outliers}")
```

```python
def cap_outliers(unseen_df, columns):
    for col in columns:
        if np.issubdtype(unseen_df[col].dtype, np.number):  # Check if the column is numeric
            Q1 = np.percentile(unseen_df[col], 25)  # 25th percentile (Q1)
            Q3 = np.percentile(unseen_df[col], 75)  # 75th percentile (Q3)
            IQR = Q3 - Q1  # Interquartile range
            
            # Define outlier bounds
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            
            # Cap the outliers
            unseen_df[col] = np.where(unseen_df[col] < lower_bound, lower_bound, unseen_df[col])
            unseen_df[col] = np.where(unseen_df[col] > upper_bound, upper_bound, unseen_df[col])
            
            print(f"Capped outliers for feature: {col}")

# Specify the columns to cap
columns_to_cap = ['views','likes', 'dislikes', 'comment_count']

# Cap the outliers in the specified columns
cap_outliers(unseen_df, columns_to_cap)
```
